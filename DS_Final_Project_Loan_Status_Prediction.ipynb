#LOAN STATUS PREDICTION

We have the data of applicants who had previously applied for the loan based on the property and applicant-based metrics.

The objective of the project is to build a Machine Learning Model to predict the loan to be approved or to be rejected for an applicant based on his/her Income, Education, Credit history and many other factors.

Feature Engineering

From the above, we can find that columns Gender, Married, Dependents, Self_Employed, LoanAmount, Loan_Amount_Term, Credit_History are having missing values. Hence it is required to impute values to make the data balanced.

```python
#Gender
```

```python
#Print the samples having null values in gender column
gender_column_null_value = train_set[train_set['Gender'].isnull()]
gender_column_null_value 
```

```python
#Impute missing values for Gender
train_set['Gender'].fillna(train_set['Gender'].mode()[0],inplace=True)
```

Feature selection

```python
X = train_set.drop(columns=['Loan_Status'],axis=1)
Y = train_set['Loan_Status']
```

```python
model = ExtraTreesClassifier()
model.fit(X,Y)
```

```python
print(model.feature_importances_)
```

```python
print("Percentage Importance of each features with respect to Loan_Status : ")
pd.Series(model.feature_importances_*100, index = train_set.columns[:-1])
```

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.plot(model.feature_importances_, train_set.columns[:-1], 'g-o', linewidth=4, markersize=10)
```

```python
prediction_features = pd.Series(model.feature_importances_*100, index = train_set.columns[:-1]).sort_values(ascending = False)
```

```python
prediction_features = prediction_features.index
```

```python
prediction_features = prediction_features[:5]
prediction_features
```

```python
print('The features that are important to loan status are')
i = 1
for j in prediction_features:
  print(i,')',j)
  i += 1
```

MODEL BUILDING

```python
feature_columns = train_set[prediction_features]
prediction_column = train_set['Loan_Status']
```

```python
X = feature_columns.values
y = prediction_column.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 3)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

--------------------------------------------------------------------------------

```python
#LOGISTIC REGRESSION
```

```python
logistic_model = LogisticRegression(max_iter = 200)
```

```python
logistic_model.fit(X_train, y_train)

print("Training Accuracy : {}%".format(logistic_model.score(X_train, y_train) * 100))
print("Testing Accuracy  : {}%".format(logistic_model.score(X_test, y_test) * 100))
```

The Logistic Regression model is having 80% Training Accuracy and 85% Testing Accuracy

```python
#DECISION TREE
```

```python
decision_tree_model = DecisionTreeClassifier(max_depth = 5) #randomly selecting the max_depth as 5
decision_tree_model.fit(X_train, y_train)
```

```python
print("Training Accuracy : {}%".format(decision_tree_model.score(X_train, y_train) * 100))
print("Testing Accuracy : {}%".format(decision_tree_model.score(X_test, y_test) * 100))
```

For max_depth of 5 we are getting Training Accuracy of 82.39% and testing accuracy of 83.11%
